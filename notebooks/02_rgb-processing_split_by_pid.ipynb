```python
import os
import shutil
from tqdm import tqdm

SRC_ROOT = "/content/drive/MyDrive/salpim_project/02_processed_frames"
DST_ROOT = "/content/drive/MyDrive/salpim_project/02_processed_frames_person_split"

# =========================
# split 정의
# =========================
train_ids = {
    "001","002","003","004","006","007","009","010",
    "012","013","014","015","016","017","019",
    "201","204","205","210","211","212","213","214",
    "217","219","220","221","222","223","224","225",
    "226","227","229","230"
}

val_ids = {
    "005","008","018","020","203","206","228"
}

test_ids = {
    "011","202","207","208","209","215","216","218"
}


# =========================
# split 찾기
# =========================
def get_split(pid):

    if pid in train_ids:
        return "train"

    elif pid in val_ids:
        return "val"

    elif pid in test_ids:
        return "test"

    return None


# =========================
# 사람 ID 추출
# =========================
def extract_person_id(video_name):

    parts = video_name.split("_")

    for p in parts:
        if p.startswith("P"):
            return p[1:]

    return None


# =========================
# 전체 video 폴더 개수 계산
# =========================
total_videos = 0

for old_split in ["train", "val", "test"]:

    split_path = os.path.join(SRC_ROOT, old_split)

    if not os.path.exists(split_path):
        continue

    for action in os.listdir(split_path):

        action_path = os.path.join(split_path, action)

        if not os.path.isdir(action_path):
            continue

        total_videos += len(os.listdir(action_path))

print(f"총 video 폴더 수: {total_videos}")


# =========================
# 복사 시작
# =========================
copied = 0
skipped = 0

pbar = tqdm(total=total_videos, desc="Copy Progress")

for old_split in ["train", "val", "test"]:

    split_path = os.path.join(SRC_ROOT, old_split)

    if not os.path.exists(split_path):
        continue

    for action in os.listdir(split_path):

        action_path = os.path.join(split_path, action)

        if not os.path.isdir(action_path):
            continue

        for video in os.listdir(action_path):

            video_path = os.path.join(action_path, video)

            if not os.path.isdir(video_path):
                pbar.update(1)
                continue

            # 사람 ID 추출
            pid = extract_person_id(video)

            if pid is None:
                skipped += 1
                pbar.update(1)
                continue

            # 새 split
            new_split = get_split(pid)

            if new_split is None:
                skipped += 1
                pbar.update(1)
                continue

            # 저장 위치
            dst_dir = os.path.join(
                DST_ROOT,
                new_split,
                action,
                video
            )

            # 이미 있으면 skip
            if os.path.exists(dst_dir):
                skipped += 1
                pbar.update(1)
                continue

            try:
                shutil.copytree(video_path, dst_dir)
                copied += 1

            except Exception as e:
                print(f"\nERROR: {video}")
                print(e)
                skipped += 1

            pbar.update(1)

            # 중간 로그
            if copied % 20 == 0:

                print("\n===================")
                print(f"copied : {copied}")
                print(f"skipped: {skipped}")
                print("===================")

pbar.close()

print("\n===== 완료 =====")
print(f"복사 성공 : {copied}")
print(f"skip/오류 : {skipped}")
```

빈 폴더만 복구

```python
import os
import shutil
from tqdm import tqdm

SRC_ROOT = "/content/drive/MyDrive/salpim_project/02_processed_frames"
DST_ROOT = "/content/drive/MyDrive/salpim_project/02_processed_frames_person_split"

# =========================
# 빈 폴더 찾기
# =========================
empty_folders = []

for split in ["train", "val", "test"]:

    split_path = os.path.join(DST_ROOT, split)

    if not os.path.exists(split_path):
        continue

    for action in os.listdir(split_path):

        action_path = os.path.join(split_path, action)

        if not os.path.isdir(action_path):
            continue

        for video in os.listdir(action_path):

            video_path = os.path.join(action_path, video)

            if not os.path.isdir(video_path):
                continue

            jpgs = [
                f for f in os.listdir(video_path)
                if f.endswith(".jpg")
            ]

            if len(jpgs) == 0:
                empty_folders.append((split, action, video))

print(f"빈 폴더 개수: {len(empty_folders)}")


# =========================
# 빈 폴더 복구
# =========================
for split, action, video in tqdm(empty_folders):

    recovered = False

    # 원본 train/val/test 전체 탐색
    for old_split in ["train", "val", "test"]:

        src_video_path = os.path.join(
            SRC_ROOT,
            old_split,
            action,
            video
        )

        if os.path.exists(src_video_path):

            dst_video_path = os.path.join(
                DST_ROOT,
                split,
                action,
                video
            )

            # 기존 빈 폴더 삭제
            shutil.rmtree(dst_video_path)

            # 다시 복사
            shutil.copytree(src_video_path, dst_video_path)

            recovered = True
            break

    if not recovered:
        print(f"복구 실패: {video}")

print("\n복구 완료")
```
